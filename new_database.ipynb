```
# clean the 'station_name' column of all_stations_df
all_stations_df1['station_name'] = all_stations_df1['station_name'].apply(lambda x : x.strip("2"))
all_stations_df1['station_name'] = all_stations_df1['station_name'].apply(lambda x : x.strip(" "))
all_stations_df1['station_name'] = all_stations_df1['station_name'].apply(lambda x : x.strip("'"))
all_stations_df1['station_name'] = all_stations_df1['station_name'].apply(lambda x : x.strip("."))

all_stations_df1.loc[3, 'station_name'] = 'Grootvlei'
all_stations_df1.loc[3, 'total_installed_capacity'] = 180
all_stations_df1.loc[6, 'station_name'] = 'Komati'
all_stations_df1.loc[28, 'station_name'] = 'Ncora Dam'

stations_table_dict = dict(zip(list(stations_table_df['station_name']), list(stations_table_df.index)))
stations_type_dict = dict(zip(list(station_type_df['station_type']), list(station_type_df.index)))
 
station_name1 = list(all_stations_df1['station_name'])
station_id = [stations_table_dict.get(a) if stations_table_dict.get(a) else a for a in station_name1]
all_stations_df1['station_id'] = station_id

station_type1 = list(all_stations_df1['station_type'])
station_type_id = [stations_type_dict.get(a) if stations_type_dict.get(a) else a for a in station_type1]
all_stations_df1['station_type_id'] = station_type_id

all_stations_df = all_stations_df1.drop(['station_name', 'station_type'], axis=1)
all_stations_df.head()
```
