```
coordinate_df = all_stations_df_updated.loc[:, ('location', 'province', 'coordinates ')]
coordinate_df.index.names = ['coordinate_id']
coordinate_df.head()
```

```
location = list(set(coordinate_df['location']))
location_df = pd.DataFrame(location, columns=['location'])
location_df.index.names = ['location_id']

location_df.head()
```

```
loc_prov_map_df = pd.DataFrame([])

province_dict = dict(zip(list(province_table_df['province']), list(province_table_df.index)))
location_dict = dict(zip(list(location_df['location']), list(location_df.index)))

province = list(coordinate_df['province'])
location1 = list(coordinate_df['location'])
prov_id = [province_dict.get(a) if province_dict.get(a) else a for a in province]
loca_id = [location_dict.get(a) if location_dict.get(a) else a for a in location1]

loca_id_new = []
for a in loca_id:
  if a == 'East London':
    loca_id_new.append(0)
  else:
    loca_id_new.append(a)

coordinate_df['province_id'] = prov_id
coordinate_df['location_id'] = loca_id_new

loc_prov_map_df['province_id'] = coordinate_df['province_id']
loc_prov_map_df['location_id'] = coordinate_df['location_id']

loc_prov_map_df.set_index(["location_id"], inplace = True)

loc_prov_map_df.head()
```

```
location_df['coordinates'] = coordinate_df['coordinates ']

location_df.head()
```

```
operator_dict = dict(zip(list(operator_df['operator']), list(operator_df.index)))
status_dict = dict(zip(list(status_df['status']), list(status_df.index)))
station_type_dict = dict(zip(list(station_type_df['station_type']), list(station_type_df.index)))

operator_list = list(all_stations_df_updated['operator '])
status_list = list(all_stations_df_updated['status '])
station_type_list = list(all_stations_df_updated['station_type'])
oper_id = [operator_dict.get(a) if operator_dict.get(a) else a for a in operator_list]
stat_id = [status_dict.get(a) if status_dict.get(a) else a for a in status_list]
stat_type_id = [station_type_dict.get(a) if station_type_dict.get(a) else a for a in station_type_list]

stat_id_new = []
for a in loca_id:
  if a == 'Decommissioned':
    stat_id_new.append(0)
  else:
    stat_id_new.append(a)

all_stations_df_updated['location_id'] = loca_id_new
all_stations_df_updated['operator_id'] = oper_id
all_stations_df_updated['status_id'] = stat_id_new
all_stations_df_updated['station_type_id'] = stat_type_id

all_stations_df_updated.head()
```
