## Data Analysis with Pandas and Python

### Import the required libraries

```
import numpy as np
import pandas as pd
from datetime import datetime 
from google.colab import files 
```

### Data

The given .bak file was converted to csv files containing the database tables. Read the csv file into **Pandas DataFrames**.

```
all_stations_df = pd.read_csv("https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.all_stations.csv")
station_type_df = pd.read_csv("https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.Station_Type.csv")
stations_table_df = pd.read_csv("https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.Stations_table.csv")
province_table_df = pd.read_csv("https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.Province%20Table%20.csv")
province_electrification_df = pd.read_csv("https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.province_electrification.csv")
coal_stations_df = pd.read_csv("https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.coal_stations%20-%20Sheet1.csv")
hydroelectric_stations_df = pd.read_csv("https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.hydroelectric_stations%20-%20Sheet1.csv")

nuclear_stations_df = pd.read_csv("https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.nuclear_stations%20-%20Sheet1.csv")
wind_stations_df = pd.read_csv("https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.wind_stations%20-%20Sheet1.csv")
gas_turbine_stations_df = pd.read_csv("https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.gas_turbine_stations%20-%20Sheet1.csv")
electricity_generated_df = pd.read_csv("https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.electricity_generated.csv")
population_df = pd.read_csv('https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.Population.csv')

year_on_year_change_df = pd.read_csv("https://raw.githubusercontent.com/Kaekaefx/Group5_Gather_Predict/master/database_tables_csv/dbo.year_on_year_change(jan_2015%20-%20jun_2019)%20-%20Sheet1.csv")
```

### Nuclear Power Stations











```
nuclear_stations_df_updated = nuclear_stations_df.copy()

# renaming column names according to specs - all lowecase snake 
nuclear_stations_df_updated = nuclear_stations_df.rename({'Power_plant':'power_plant','Province':'province', 'Coordinates':'coordinates', 'Status	': 'status','_Capacity__Megawatt__': 'capacity_MW', '_Date_commissioned_' : 'date_commissioned', '_Annual_output_GWh_':'annual_output_MWh' , 'Capacity_factor__':'capacity_factor'}, axis=1)

# changing from GWh to MWh
nuclear_stations_df_updated['annual_output_MWh'] = nuclear_stations_df_updated['annual_output_MWh']/1000 

# updated table 
nuclear_stations_df_updated.head()

nuclear_stations_df_updated.to_csv(r'nuclear_stations_df_updated')
files.download('nuclear_stations_df_updated')
```

### Hydroelectric Stations

```
hydroelectric_stations_df_updated = hydroelectric_stations_df.copy()

# renaming column names according to specs - all lowecase snake 
hydroelectric_stations_df_updated = hydroelectric_stations_df_updated.rename({'Power_plant':'power_plant','Province':'province', 'Coordinates':'coordinates', 'Status': 'status','Operator': 'operator', 'Notes' : 'notes', '_Installed_capacity__planned___Megawatt__':'installed_capacity_planned_MW' , '_Date_commissioned__planned__':'date_commissioned_planned'}, axis=1)

# renaming entries in the province column that have brackets 
hydroelectric_stations_df_updated['province'][0] = 'Limpopo'
hydroelectric_stations_df_updated['province'][9] = 'Free State'

# Getting the commission dates all in the same format
hydroelectric_stations_df_updated['date_commissioned_planned'][0] = np.NaN
hydroelectric_stations_df_updated['date_commissioned_planned'][11] = 2015

# updated table 
hydroelectric_stations_df_updated.head(20)

# should we find the missing information and fill it in or not? 
# Griep dam goes over 2 provinces - issues with third normal form 

nuclear_stations_df_updated.to_csv(r'hydroelectric_stations_df_updated')
files.download('hydroelectric_stations_df_updated')
```

### Gas Turbine Stations

```
# renaming column names according to specs - all lowecase snake 
gas_turbine_stations_df_updated = gas_turbine_stations_df.rename({'Power_plant':'power_plant','Province':'province', 'Coordinates':'coordinates', 'Status': 'status','Operator': 'operator', 'Notes' : 'notes', '_Installed_capacity__Megawatt__':'installed_capacity_planned_MW' , '_Date_commissioned_':'date_commissioned'}, axis=1) 

# updated table 
gas_turbine_stations_df_updated.head()

# creating new csv file
nuclear_stations_df_updated.to_csv(r'gas_turbine_stations_df_updated')
files.download('gas_turbine_stations_df_updated')
```

### Wind Stations 

```
wind_stations_df_updated = wind_stations_df.copy()

# Rename columns - lowercase snake
wind_stations_df_updated = wind_stations_df_updated.rename({'Coordinates':'coordinates', 'Status': 'status','Operator': 'operator', 'Notes' : 'notes', 'Capacity_factor____expected_':'capacity_factor_expected', '_Annual_output_GWh__expected__': 'annual_output_MWh_expected', '_Capacity_MW__planned__': 'capacity_MW_planned', '_Date_commissioned__planned__':'date_commissioned_planned', 'Province':'province', 'Power_plant':'power_plant'}, axis = 1)

# remove extra info in brackets in province column entries 
wind_stations_df_updated['province'][0] = 'Western Cape'
wind_stations_df_updated['province'][1] = 'Western Cape'

# removing extra info in 'status' column entries 
wind_stations_df_updated['status'][0] = 'Decommissioned'

# date_commissioned_planned column needs attention !!! 

# Adding a 'notes' column 
wind_stations_df_updated['notes'] = ['Prototype/Research', np.NaN]

# updated table 
wind_stations_df_updated.head()

# creating new csv file
nuclear_stations_df_updated.to_csv(r'wind_stations_df_updated')
files.download('wind_stations_df_updated')
```

### Electricity Generated 

```
# create a copy of the table to edit 
electricity_generated_df_updated = electricity_generated_df.copy()

# get rid of the timestamp 
electricity_generated_df_updated['Date(mm-yyyy)'] = [item[:10] for item in electricity_generated_df_updated['Date(mm-yyyy)']]

# Rename the date column 
electricity_generated_df_updated = electricity_generated_df_updated.rename({'Date(mm-yyyy)':'date'}, axis = 1)

# Change the data type to float in the electricity_distributed column by removing the commas in the numbers 
electricity_generated_df_updated = electricity_generated_df_updated.apply(lambda x: x.str.replace(',','.'))
electricity_generated_df_updated["electricity_distributed"] = pd.to_numeric(electricity_generated_df_updated["electricity_distributed"])

# Make sure the date column has the correct data type 
# electricity_generated_df_updated['date'][0] = datetime.strptime(electricity_generated_df_updated['date'][0], '%Y-%m-%d').date()
# print(type(electricity_generated_df_updated['date'][0]))

# updated table 
electricity_generated_df_updated.head(20) 

# creating new csv file
nuclear_stations_df_updated.to_csv(r'electricity_generated_df_updated')
files.download('electricity_generated_df_updated')
```
